```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('kc_house_data.csv')
```

```python
df.head()
```

#### 1.2 Изучите стоимости недвижимости

```python
plt.figure(figsize = (6,4))
plt.hist(df['price'], bins = 10)
plt.xlabel('Стоимость')
plt.ylabel('Количество')
plt.title('Стоимость недвижимости');
```

#### 1.3 Изучите распределение квадратуры жилой

```python
plt.figure(figsize = (6,4))
sns.histplot(df['sqft_living'], bins = 10)
plt.xlabel('Жилые квадратные метры')
plt.ylabel('Количество')
plt.title('Распределение жилой квадратуры');
```

#### 1.4 Изучите распределение года постройки

```python
plt.figure(figsize = (6,4))
sns.histplot(df['yr_built'], bins = 20)
plt.xlabel('Год постройки')
plt.ylabel('Количество')
plt.title('Распределение года постройки');
```

#### 2.1 Изучите распределение домов от наличия вида на набережную<br>
<li>Постройте график
<li>Сделайте выводы<br>

```python
view = df['waterfront'].value_counts()

plt.figure(figsize = (6,4))
plt.pie(view.values, autopct = '%1.1f%%')
plt.title('Распределение домов от наличия вида на набережную')
plt.legend(['No', 'Yes']);
```

*Вид на набережную имеют лишь 0,8% от Всех домов*

#### 2.2 Изучите распределение этажей домов

```python
floors = df['floors'].value_counts()

plt.figure(figsize = (6,4))
plt.pie(floors.values, autopct = '%1.1f%%', labels = floors.index)
plt.title('Распределение этажей домов');
```

*Большая часть домов состоит из 1 и 2 этажей. Одноэтажные дома лидируют.*

#### 2.2 Изучите распределение состояния домов

```python
condition = df['condition'].value_counts()

plt.figure(figsize = (6,4))
plt.pie(condition.values, autopct = '%1.1f%%', labels = condition.index)
plt.title('Распределение состояния домов');
```

*Наибольшее кол-во домов имеют оценку по состоянию дома - 3. Таких домов 64,9%.*

## Задача 3
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.<br>
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
corr_matrix = df.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) <= 0.3] = 0
corr_matrix
```

```python
plt.figure(figsize = (8,6))
sns.heatmap(corr_matrix, annot = True);
```

```python
sns.jointplot(x = df['price'], y = df['grade'], kind = 'reg');
```

*Стоимость дома становится выше, когда качество конструкции и дизайн дома повышаются*

```python
sns.jointplot(x = df['price'], y = df['sqft_living'], kind = 'reg');
```

*Стоимость дома становится выше, когда жилая площадь больше*

```python
plt.figure(figsize = (6,4))
sns.boxplot(x = df['price'], y = df['floors'].astype('str'), whis = 1.5)
plt.xlabel('price')
plt.ylabel('floors')
plt.title('Стоимость/Этажность дома');
```

*Дома, которые имеют 2,5 этажа - имеют наибольшую стоимость*

```python
plt.figure(figsize = (6,4))
sns.boxplot(x = df['price'], y = df['view'].astype('str'), whis = 1.5)
plt.xlabel('price')
plt.ylabel('view')
plt.title('Стоимость/Оценка вида');
```

*Дома, у которых оценка вида = 4 - имеют наибольшую стоимость*

```python
plt.figure(figsize = (6,4))
sns.boxplot(x = df['price'], y = df['grade'].astype('str'), whis = 1.5)
plt.xlabel('price')
plt.ylabel('grade')
plt.title('Стоимость/Оценка конструкции и дизайна');
```

*Дома, у которых оценка конструкции и дизайна = 13 - имеют наибольшую стоимость*
